```python
def convertToList(genres):
    genresList = []
    for genre in ast.literal_eval(genres):
        genresList.append(genre['name'])
    return genresList
```

```python
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
movies['genres'] = movies['genres'].apply(convertToList)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convertToList)
```

```python
movies.head(1)
```

```python
def getThreeCast(casts):
    castList = []
    count = 0
    for cast in ast.literal_eval(casts):
        if count == 3:
            break
        castList.append(cast['name'])
        count += 1
    return castList
```

```python
movies['cast'] = movies['cast'].apply(getThreeCast)
```

```python
movies.head(1)
```

```python
def fetchDirector(crews):
    directorsList = []
    for crew in ast.literal_eval(crews):
        if crew['job'] == 'Director':
            directorsList.append(crew['name'])
    return directorsList
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def removiewhiteSpace(contentList):
    i=0
    for content in contentList:
        crew = content.replace(" ","")
        contentList[i] = content
        i +=1
    return contentList
```

```python
movies['genres'] =movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] =movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] =movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(removiewhiteSpace)
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
newMovieDataFrame = movies[['movie_id','title','tags']]
```

```python
newMovieDataFrame.head()
```

```python
newMovieDataFrame.loc[:,'tags'] = newMovieDataFrame.loc[:,'tags'].apply(lambda x:" ".join(x))
```

```python
newMovieDataFrame['tags'] = newMovieDataFrame['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    stemmedTextList = []
    
    for i in text.split():
        stemmedTextList.append(ps.stem(i))
    return " ".join(stemmedTextList)
```

```python
newMovieDataFrame['tags'] = newMovieDataFrame['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000, stop_words = "english")
```

```python
featureVectors = vectorizer.fit_transform(newMovieDataFrame['tags']).toarray()
```

```python
featureVectors[0]
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(featureVectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])
```

```python
def recommend(movie):
    movieIndex = newMovieDataFrame[newMovieDataFrame['title'] == movie].index[0]
    distances = similarity[movieIndex]
    moviesList = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in moviesList:
        print(newMovieDataFrame.iloc[i[0]].title)
```

```python
recommend('Predators')
```
